# Check whether all the zones generate the same amount of sales

```python
#Null hypothesis and alternate hypothesis
#H0: All the zones generate the same amount of sales
#Ha: All the zones generate different amount of sales
#alpha = 0.05
```

```python
from scipy.stats import f_oneway
f_oneway(df['Zone - A'], df['Zone - B'], df['Zone - C'], df['Zone - D'])
```

```python
#p value obtained is less than alpha. So we reject the null hypothesis. Thus, all the zones generate different amount of sales.
```
